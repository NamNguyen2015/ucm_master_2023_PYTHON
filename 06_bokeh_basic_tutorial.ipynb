```python
# Standard imports 

from bokeh.io import output_notebook, show
output_notebook()
```

```python
# Plot a complex chart with interactive hover in a few lines of code

from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.sampledata.autompg import autompg_clean as df
from bokeh.transform import factor_cmap

```

```python
df.describe(include = 'all')
```

```python
df.head()
```

```python
df.cyl = df.cyl.astype(int)
df.yr = df.yr.astype(int)
```

```python
df.describe()
```

```python
df.cyl = df.cyl.astype(str)
df.yr = df.yr.astype(str)

group = df.groupby(by=['cyl', 'mfr'])
```

```python
group.head()
```

```python
source = ColumnDataSource(group)
```

```python
source
```

```python
p = figure(width=800, 
           height=300, 
           title="Mean MPG by # Cylinders and Manufacturer",
           x_range=group, 
           toolbar_location=None, 
           tools="")

p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Manufacturer grouped by # Cylinders"
p.xaxis.major_label_orientation = 1.2
```

```python
show(p)
```

```python

index_cmap = factor_cmap('cyl_mfr', 
                         palette=['#2b83ba', 
                                  '#abdda4', 
                                  '#ffffbf', 
                                  '#fdae61', 
                                  '#d7191c'], 
                         factors=sorted(df.cyl.unique()), 
                         end=1)

```

```python
p.vbar(x='cyl_mfr', 
       top='mpg_mean', 
       width=1, 
       source=source,
       line_color="white", 
       fill_color=index_cmap, 
       hover_line_color="darkgrey", 
       hover_fill_color=index_cmap)

```

```python
show(p)
```

```python
p.add_tools(HoverTool(tooltips=[("MPG", "@mpg_mean"), 
                                ("Cyl, Mfr", "@cyl_mfr")]))

show(p)
```
